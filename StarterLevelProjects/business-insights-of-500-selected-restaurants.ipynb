# **Data Exploration**

### Cuisines

```python
import matplotlib.pyplot as plt
import seaborn as sbr
```

```python
# PROCESS CUISINES (Split and Explode)
    
cuisine_df = df.assign(Cuisines=df['Cuisines'].str.split(', ')).explode('Cuisines')

# ALWAYS RESET THE INDEX AFTER .agg()

cuisine_stats = cuisine_df.groupby('Cuisines').agg(count=('Restaurant ID', 'count'), average_rating=('Aggregate rating', 'mean')).reset_index()
                                    
```

```python
# PLOTTING AGGREGATED DATA

top_20_cuisines = cuisine_stats.sort_values(by='count', ascending=False).head(20)

plt.figure(figsize=(8,6))

sbr.barplot(x='Cuisines', y='count', data=top_20_cuisines, palette='Paired')

plt.title('Top 20 Most Common Cuisines')
plt.xlabel('Cuisine')
plt.xticks(rotation=45)
plt.ylabel('Number of Restaurants')
plt.tight_layout()
plt.show()
```

```python
ratings_counts = cuisine_df.groupby('Cuisines').agg(average_rating=('Aggregate rating', 'mean')).reset_index()
```

```python

top_20_ratings = cuisine_stats[cuisine_stats['count'] > 100].sort_values(by='average_rating', ascending=False).head(20)

plt.figure(figsize=(8,6))

sbr.barplot(x='average_rating', y='Cuisines', data=top_20_ratings, palette='flare')

plt.title('Top 20 Best Rated Cuisines (min. 100 Ratings)')
plt.xlabel('Average Rating')
plt.xticks(rotation=45)
plt.ylabel('Cuisines')
plt.tight_layout()
plt.show()
```

## Currencies

```python
plt.figure(figsize=(10, 6))

top_10_currencies = df['Currency'].value_counts().nlargest(5).plot(kind='barh', color=plt.cm.Paired.colors)

plt.title('Restaurants by Currencies')
plt.tight_layout()
plt.grid()
plt.show()
```

## Price Ranges

```python

plt.figure(figsize=(8, 6))
price_ranges = df['Price range'].value_counts()
plt.pie(
    price_ranges,
    labels=price_ranges.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.tab10.colors
)
plt.title('% of Restaurants by Price Range (1: Low to 4: V High)')
plt.tight_layout()
plt.show()
```

```python
df.groupby("Price range")['Aggregate rating'].mean().reset_index()

plt.figure(figsize=(8,6))

sbr.violinplot(x='Price range', y='Aggregate rating', data=df, palette='flare')

plt.title('Average Rating by Price Range')
plt.xlabel('Price Range (1: Low to 4: V High)')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

## Engagement

```python
plt.figure(figsize=(8,5))

price_engagement = df.groupby('Price range')['Votes'].mean().reset_index()
sbr.lineplot(x="Price range", y="Votes", data=price_engagement, marker='o', linewidth=2.5, color='orange')

plt.title("Average Restaurant Votes By Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Votes")
plt.tight_layout()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4,4))

booking_engagement = df.groupby('Has Table booking')['Votes'].mean().reset_index()
sbr.barplot(x="Has Table booking", y="Votes", data=booking_engagement, palette='Greens_d')

plt.title("Engagement vs Table Bookings")
plt.xlabel("Table Bookings")
plt.ylabel("Average Votes")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4,4))

delivery_engagement = df.groupby('Has Online delivery')['Votes'].mean().reset_index()
sbr.barplot(x="Has Online delivery", y="Votes", data=delivery_engagement, palette='Greens_d')

plt.title("Engagement vs Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Average Votes")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

sbr.histplot( 
    data=df["Aggregate rating"],
    bins=50,
    kde=True
)

plt.title("Distribution of Ratings")
plt.tight_layout()
plt.show()
```
